```python
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df_tita = df[['Survived','Pclass','Sex','Age','Embarked']]
```

```python
df_tita = df_tita.dropna()
```

```python
df_tita.head()
```

```python
len(df_tita)
```

```python
#성별, 객실등급별 승선자 수
df_tita.pivot_table(index='Sex',columns='Pclass',values='Survived',aggfunc='count',margins=True)
```

```python
#성별, 객실등급별 생존자 수
df_tita.pivot_table(index='Sex',columns='Pclass',values='Survived',aggfunc='sum',margins=True)
```

```python
#성별, 객실등급별 생존율
df_tita.pivot_table(index='Sex',columns='Pclass',values='Survived',aggfunc='mean',margins=True)
```

```python
df = df[['Survived','Pclass','Sex','Age','Embarked']]
df = df_tita.dropna()
```

```python
len(df)
```

```python
#객실등급(Pclass) 승선자 수를 구한 결과를 데이터프레임으로 만들기
df1 = df.groupby('Pclass').Survived.count().to_frame()
df1
```

```python
#객실등급(Pclass) 생존자 수를 구한 결과를 데이터프레임으로 만들기
df2 = df.groupby('Pclass').Survived.sum().to_frame()
df2
```

```python
#객실등급(Pclass) 생존율을 구한 결과를 데이터프레임으로 만들기
df3 = df.groupby('Pclass').Survived.mean().to_frame()
df3
```

```python
#객실등급별 탑승자수 생존자수 생존율 데이터프레임 만들기
df4 = pd.concat([df1,df2,df3],axis=1)
df4.columns = ['승선자수','생존자수','생존율']
df4
```

```python
#성별 탑승자 수 데이터프레임 만들기
df5 = df.groupby('Sex').Survived.count().to_frame()
df5
```

```python
#성별 생존자 수 데이터프레임 만들기
df6 = df.groupby('Sex').Survived.sum().to_frame()
df6
```

```python
#성별 생존율 데이터프레임 만들기
df7 = df.groupby('Sex').Survived.mean().to_frame()
df7
```

```python
#성별 탑승자수 생존자수 생존율 데이터프레임 만들기
df8 = pd.concat([df5,df6,df7],axis=1)
df8.columns = ['승선자수','생존자수','생존율']
df8
```

```python
df.groupby(['Pclass','Sex']).Survived.mean()
```

```python
#함수 적용하기
def my_mean(values):
  return sum(values)/len(values)
```

```python
#함수 적용하기
df.groupby(['Pclass','Sex']).Survived.agg(my_mean)
```
