```python
df = pd.read_csv('/content/서울시 코로나19 확진자 현황.csv')
```

```python
df
```

```python
df['환자번호'].unique()
```

```python
#불필요한 컬럼 삭제
df.drop(columns=['환자번호','국적','환자정보','조치사항','이동경로','등록일','수정일','노출여부'],axis=1,inplace = True)
```

```python
df
```

```python
#자료형 확인
df.dtypes
```

```python
#자료형 변환
df['확진일'] = pd.to_datetime(df['확진일'])
```

```python
df['지역'].unique()
```

```python
df['지역'] = df['지역'].str.strip()
```

```python
df['지역'].nunique()
```

```python
df['지역'] = df['지역'].astype('category')
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_gu = pd.pivot_table(df, index = '확진일' , columns = '지역', values = '연번', aggfunc ='count', margins = True)
```

```python
#확진일, 지역(구)별 확진자 수
df_gu
```

```python
s_date = df_gu['All'][:-1]
s_date
```

```python
s_date.sort_values(ascending=False)
```

```python
x= s_date.index
y= s_date.values
plt.title('Seoul')
plt.plot(x,y)
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.show()
```

```python
df_gu
```

```python
#서울시 구별 누적확진자 많은순으로 보기
s_gu = df_gu.loc['All'][:-1]
```

```python
s_gu
```

```python
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
x= s_gu.index
y= s_gu.values
plt.title('Seoul')
plt.bar(x,y)
plt.xlabel('region')
plt.xticks(rotation=45)
plt.show()
```

```python
#최근일 기준 지역별 추가 확진자
s_gu = df_gu.iloc[-2][:-1]
s_gu.sort_values(ascending=False)
```
