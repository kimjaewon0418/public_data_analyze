```python
import pandas as pd
```

```python
df = pd.read_csv('/content/scores.csv')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['avg'] = round((df.kor+df.eng+df.math)/3,1)
df
```

```python
def get_grade(x):
  if x >= 90:
    return 1
  elif x >= 80:
    return 2
  elif x >= 70:
    return 3
  elif x>=60:
    return 4
  else:
    return 5

df['grade'] = df['avg'].apply(get_grade)
df.head(3)
```

```python
df.dtypes
```

```python
df['grade'] = df['grade'].astype('category')
```

```python
df['grade'].dtypes
```

```python
df['grade'] = df['grade'].cat.rename_categories(['A', 'B', 'C', 'D'])
df
```

```python
df['grade'] = df['grade'].cat.set_categories(['A', 'B', 'C', 'D', 'F'])
df['grade'].dtypes
```
